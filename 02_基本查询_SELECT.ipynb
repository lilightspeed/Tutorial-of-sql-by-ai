## 1. SELECT 基本语法

```sql
SELECT 列1, 列2, ...
FROM 表名
WHERE 条件;
```

执行顺序：
1. `FROM` — 确定从哪张表取数据
2. `WHERE` — 过滤符合条件的行
3. `SELECT` — 选择要显示的列

````python
import pymysql
import pandas as pd

DB_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "your_password",
    "database": "shop_db",
    "charset": "utf8mb4",
}

def run_query(sql, params=None):
    conn = pymysql.connect(**DB_CONFIG)
    try:
        return pd.read_sql(sql, conn, params=params)
    finally:
        conn.close()
````

## 2. 查询指定列

只选择你需要的列，而不是 `*`（所有列）。

```sql
SELECT 列名1, 列名2 FROM 表名;
```

````python
# 查询客户的姓名和城市
run_query("SELECT customer_name, city FROM customers")
````

````python
# 查询商品名称和价格
run_query("SELECT product_name, price FROM products")
````

## 3. 查询所有列

使用 `*` 代表所有列（实际开发中建议明确列出需要的列）。

```sql
SELECT * FROM 表名;
```

````python
# 查询所有商品信息
run_query("SELECT * FROM products")
````

## 4. WHERE 条件过滤

WHERE 子句用于过滤满足条件的行。

```sql
SELECT 列名 FROM 表名 WHERE 条件;
```

### 常用比较运算符

| 运算符 | 说明 | 示例 |
|--------|------|------|
| `=` | 等于 | `city = '北京'` |
| `!=` 或 `<>` | 不等于 | `status != '已完成'` |
| `>` | 大于 | `price > 100` |
| `<` | 小于 | `price < 50` |
| `>=` | 大于等于 | `stock >= 100` |
| `<=` | 小于等于 | `salary <= 15000` |

````python
# 查询北京的客户
run_query("SELECT * FROM customers WHERE city = '北京'")
````

````python
# 查询价格大于 1000 的商品
run_query("SELECT product_name, price FROM products WHERE price > 1000")
````

````python
# 查询已完成的订单
run_query("SELECT order_id, order_date, total_amount FROM orders WHERE status = '已完成'")
````

## 5. DISTINCT 去重

`DISTINCT` 用于消除重复行，只返回唯一的值。

```sql
SELECT DISTINCT 列名 FROM 表名;
```

````python
# 查询客户所在的城市（去重）
run_query("SELECT DISTINCT city FROM customers")
````

````python
# 查询商品的分类ID（去重）
run_query("SELECT DISTINCT category_id FROM products")
````

## 6. LIMIT 限制行数

`LIMIT` 用于限制返回的记录数量，常用于分页或预览数据。

```sql
SELECT 列名 FROM 表名 LIMIT N;        -- 返回前 N 行
SELECT 列名 FROM 表名 LIMIT N OFFSET M; -- 跳过前 M 行，返回 N 行
```

````python
# 只查看前 3 个商品
run_query("SELECT product_name, price FROM products LIMIT 3")
````

````python
# 分页：跳过前 3 条，取 3 条（第 4-6 条）
run_query("SELECT product_name, price FROM products LIMIT 3 OFFSET 3")
````

## 7. 别名 AS

使用 `AS` 为列或表设置别名，让输出更易读。

```sql
SELECT 列名 AS 别名 FROM 表名 AS 表别名;
```

````python
# 为列设置中文别名
run_query("""
    SELECT 
        product_name AS 商品名称,
        price AS 价格,
        stock AS 库存
    FROM products
    LIMIT 5
""")
````

````python
# 在表达式中使用别名
run_query("""
    SELECT 
        product_name AS 商品名称,
        price AS 单价,
        stock AS 库存数量,
        price * stock AS 库存总价值
    FROM products
    LIMIT 5
""")
````

## 8. 计算列

SELECT 中可以使用算术运算符对列进行计算。

| 运算符 | 说明 | 示例 |
|--------|------|------|
| `+` | 加 | `price + 10` |
| `-` | 减 | `price - discount` |
| `*` | 乘 | `price * quantity` |
| `/` | 除 | `total / count` |

````python
# 计算每个商品打 8 折后的价格
run_query("""
    SELECT 
        product_name AS 商品名称,
        price AS 原价,
        price * 0.8 AS 折后价
    FROM products
    LIMIT 5
""")
````
